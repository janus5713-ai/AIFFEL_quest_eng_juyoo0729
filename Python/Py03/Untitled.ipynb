사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다
정수가 입력될 때 까지 "잘못된 입력입니다. 정수를 입력해주세요."를 출력하며 다시 입력받기를 시도합니다
나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다
사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다
math 모듈을 사용하여 제곱 연산(**)을 지원합니다
계산기를 통해 계속 계산을 할 것인지 입력받습니다


```python
import math  # 제곱 연산을 위해 math 모듈 불러오기

# 계속 계산하기 위한 반복문
while True:

    # 첫 번째 정수 입력
    # 힌트: 정수 입력될 때까지 반복 (while True + break)
    while True:
        try:
            # 힌트: int(input()) 으로 정수 입력받기
            num1 = int(input("첫 번째 정수를 입력하세요: "))
            break
        except ValueError:
            # 힌트: "잘못된 입력입니다. 정수를 입력해주세요." 출력
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    # 힌트: input() 으로 연산자 입력받기 (+,-,*,/,**)
    operator = input("연산자를 입력해세요 (+,-,*,/,**): ")

    # 두 번째 정수 입력
    # 힌트: 첫 번째 정수 입력이랑 똑같이!
    while True:
        try:
            num2 = int(input("두 번째 정수를 입력하세요: "))
            break
        except ValueError:
            print("잘못된 입력입니다. 정수를 입력해주세요.")

    # 연산자에 따라 계산하기
    try:
        # 힌트: if/elif 로 각 연산자 처리
        if operator == '+':
            print(f"결과: {num1 + num2}")       # 힌트: num1 + num2
        elif operator == '-':
            print(f"결과: {num1 - num2}")       # 힌트: num1 - num2
        elif operator == '*':
            print(f"결과: {num1 * num2}")       # 힌트: num1 * num2
        elif operator == '/':
            print(f"결과: {num1 / num2}")       # 힌트: num1 / num2
        elif operator == '**':
            print(f"결과: {math.pow(num1, num2)}")       # 힌트: math.pow() 사용!
        else:
            # 힌트: 지원하지 않는 연산자 메시지 출력
            print("지원하지 않는 연사자입니다")

    # 힌트: 0으로 나누기 오류 잡기
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다")                      # 힌트: 0으로 나눌 수 없다는 메시지

    # 계속 계산할지 물어보기
    # 힌트: input() 으로 y/n 입력받기
    again = input("계속 계산합니까? Y/N:")
    if again == 'n':
        print("계산이 종료되었습니다")                      # 힌트: 종료 메시지
        break                            # 힌트: 반복문 탈출!
```
